```python
import cv2
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# grayscale and thresholding
imageName = "2.jpg"
img = cv2.imread("../data/inputs/"+imageName)
# img = cv2.imread("2.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=[15,15])
plt.title('Original Image')
plt.imshow(img_gray,cmap="gray")
plt.show()

value,thresholded_img = cv2.threshold(img_gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.figure(figsize=[15,15])
plt.title('Thresholded Image')
plt.imshow(thresholded_img,cmap="gray")
plt.show()

print(f'Threshold value: {value}')
```

```python
# inversion
inv = 255-thresholded_img
plt.figure(figsize=[15,15])
plt.title('Inverted Image')
plt.imshow(inv,cmap="gray")
plt.show()
```

```python
# erosion
kernel = np.ones((1,100))
kernel = kernel.astype(np.uint8)
erosion = cv2.erode(inv,kernel,iterations = 3)
plt.figure(figsize=[20,20])
plt.title('Eroded Image')
plt.imshow(erosion, cmap='gray')
plt.show()
```

```python
# get full lines 
reduced_width = cv2.reduce(erosion,1, cv2.REDUCE_AVG).reshape(-1)
h, w = img.shape[:2]
horizontal_lines = np.zeros((h,w))

for row in range(h):
    if reduced_width[row]!=0:
        cv2.line(horizontal_lines, (0,row), (w,row), (255), 1)

plt.figure(figsize=[20,20])
plt.title('Stave lines')
plt.imshow(horizontal_lines,cmap='gray')
plt.show()
```

```python
# subtraction
unsigned_lines = horizontal_lines.astype(np.uint8)
notes = cv2.subtract(inv, unsigned_lines)
plt.figure(figsize=[20,20])
plt.title('Subtracted Image')
plt.imshow(notes,cmap='gray')
plt.show()
```

```python
# closing
kernel_31 = np.ones((3,1))
kernel_31 = kernel_31.astype(np.uint8)

kernel_41 = np.ones((4,1))
kernel_41 = kernel_41.astype(np.uint8)

inv_notes_closed = cv2.dilate(notes, kernel_31)
inv_notes_closed = cv2.morphologyEx(inv_notes_closed, cv2.MORPH_OPEN, kernel_41)
inv_notes_closed = cv2.erode(inv_notes_closed, kernel_41)

plt.figure(figsize=[20,20])
plt.title('Only notes')
plt.imshow(inv_notes_closed, cmap='gray')
plt.show()
```

```python
# segmenting image into parts
kernel = np.ones((40,1))
kernel = kernel.astype(np.uint8)
dilated_lines = cv2.dilate(horizontal_lines, kernel, iterations = 1)
plt.figure(figsize=[15,15])
plt.title('Dilated Image')
plt.imshow(dilated_lines,cmap='gray')
plt.show()
```

```python
hist = cv2.reduce(dilated_lines, 1, cv2.REDUCE_AVG).reshape(-1)

bool, start = 0, 0
H,W = img.shape[:2]
red_lines=[]

for y in range(H-1):
    if y==H-2:
        if bool == 0:
            row_no = start+int((y-start)/2)
            red_lines.append(row_no)
            bool = 1
    else:    
        if hist[y] == 255:
            if bool == 0:
                row_no = start+int((y-start)/2)
                red_lines.append(row_no)
                bool = 1
        elif bool == 1 and hist[y] == 0:
            bool = 0
            start = y
        
red_image = thresholded_img
red_image = cv2.cvtColor(red_image, cv2.COLOR_GRAY2BGR)

color = (255,0,0)
thickness = 1
for y in red_lines:
    cv2.line(red_image, (0,y), (W, y), color, thickness)

plt.figure(figsize=[20,20])
plt.title('Red Lines')
plt.imshow(red_image)
plt.show()
```

```python
def getConsecutiveIntervals(lst):
    if(len(lst)<1):
        return []
    intervals = []
    start = 0
    end = 0
    for i in range(1,len(lst)):
        if lst[i] == (lst[i-1]+1):
            end = i
        else:
            intervals.append([start,end])
#             print(lst[start],lst[end])
            start = i
            end = i
    intervals.append([start,end])
    return intervals

def group(lst):
    start = 0
    end = 0
    for i in range(1,len(lst)):
        if lst[i] != (lst[i-1]+1):
            yield(lst[start],lst[end])
            start = i
            end = i
        else:
            end = i
    yield(lst[start],lst[end])  
    
def display_images(images, rows, cols, size, titles):
    fig = plt.figure(figsize = size)
    c = 0
    size = len(images)
    for i in range(rows):
        for j in range(cols):
            if size == c:
                break
            ax = fig.add_subplot(rows, cols, c + 1)
            ax.imshow(images[c], cmap = 'gray')
            ax.set_title(titles[c], fontsize = 20)
            c = c + 1
    plt.show()
```

**Quarter Note Detection**

```python
def circle_SE(r):
    length = 2*r+1
    SE = np.zeros((length,length))
    for i in range(length):
        for j in range(length):
            if((i-r)*(i-r) + (j-r)*(j-r) <= r*r):
                SE[i,j]=1
    return np.uint8(SE)
    
print(circle_SE(5))
```

**Cleft Detection**

```python
stave_rows = []
for i in range(3):
    stave_rows.append(staves[int(len(staves)/3)*i])

print(f'First 3 rows: {stave_rows}')
```

```python
cols = [-1, -1, -1]
yellow = [242, 237, 106]
bw_notes = inv_notes_closed.copy()

top_corners = []

for col in range(bw_notes.shape[1]):
    for i in range(3):
        if cols[i]==-1 and bw_notes[stave_rows[i]][col]==255:
            cols[i] = col
            top_corners.append([stave_rows[i], cols[i]])
    if cols[0]!=-1 and cols[1]!=-1 and cols[2]!=-1:
        break

dx = [-1,1,0,0]
dy = [0,0,1,-1]

def dfs(point):
    row,col = point[0], point[1]
    for i in range(4):
        nrow, ncol = row+dx[i], col+dy[i]
        if bw_notes[nrow][ncol] != 0:
            bw_notes[nrow][ncol] = 0
            notes[nrow][ncol][:] = yellow[:]
            dfs([nrow,ncol])
    
for point in top_corners:
    dfs(point)
    
kernel = np.ones((1,2))
neg_array = cv2.morphologyEx(255-inv, cv2.MORPH_CLOSE, kernel)
region_cnt = len(red_lines)
cleft = np.zeros(region_cnt-1)

for j in range(region_cnt):
    if j+1<region_cnt:
        region = neg_array[red_lines[j]:red_lines[j+1],:]
        region = 255 - region
        sum1 = np.sum(region, axis=0)
        mode_val = stats.mode(sum1[sum1!=0])
        mode_val = mode_val[0][0]
        idx = np.where(sum1!=mode_val)
        idx_list = list(group(idx[0]))
        idx_list = idx_list[1]        
        if(np.sum(inv[staves[(j+1)*5-1]+5: red_lines[j+1], idx_list[0]: idx_list[1]])>0):
            cleft[j]=0
        else:
            cleft[j]=1

plt.figure(figsize=[10,10])
plt.title('Clefts marked in yellow')
plt.imshow(notes,cmap='gray')
plt.show()
```

**1/8th Note Detection**

```python
def vertical_element(x):
    s = (x,x)
    arr = np.zeros(s,dtype=np.uint8)
    for i in range(0,x):
        for j in range(0,x):
            if j == x//2:
                arr[i][j] = 1
    return arr
def horizontal_element(x):
    s = (x,x)
    arr = np.zeros(s,dtype=np.uint8)
    for i in range(0,x):
        for j in range(0,x):
            if i == x//2:
                arr[i][j] = 1
    return arr
```

```python
def diagonal_element(x,w,i): ## i = 1 for inverted
    s = (x-i*w,x-i*w)
    arr = np.zeros(s,dtype=np.uint8)
    for i1 in range(0,x-w):
        for j in range(0,x-w*i):
            if j == x//2:
                arr[i1][j] = 1
    var = 0
    while var < w:
        arr[var,x//2+var] = 1
        if i == 0:
            arr[x-w+var,x//2-var] = 1
        if i == 1:
            arr[-var-1,x//2+var] = 1
        var = var + 1
    return arr
```

```python
def oneEighthDetection(x,xv1,xv2,ival):
    var1 = len(red_lines)
    red = [x,0,0]
    i = 0
    
    while i < var1 - 1:

        s1 = (red_lines[i+1]-red_lines[i],inv.shape[1])
        notes_row = np.zeros(s1)
        for i1 in range(0,notes_row.shape[0]):
            for j1 in range(0,notes_row.shape[1]):
                notes_row[i1][j1] = inv[i1+red_lines[i]][j1]
                   
        s2 = (red_lines[i+1]-red_lines[i],notes.shape[1],notes.shape[2])
        notes3d_row = np.zeros(s2)
                
        for i1 in range(0,notes3d_row.shape[0]):
            for j1 in range(0,notes3d_row.shape[1]):
                for z1 in range(0,notes3d_row.shape[2]):
                    notes3d_row[i1][j1][z1] = notes[i1+red_lines[i]][j1][z1]
                    
        varb1 = staves[3]-staves[0]
        
        if ival == 0:
            morphed_row=cv2.morphologyEx(notes_row, cv2.MORPH_OPEN, diagonal_element(varb1,8,0))
        else:
            morphed_row=cv2.morphologyEx(notes_row, cv2.MORPH_OPEN, diagonal_element(varb1,8,1))
            
        varval = 255
        indcs=np.where(morphed_row == varval)
        indc_len = len(indcs[0])
        if(indc_len != 0):
            indcs1=np.where(np.sum(morphed_row,axis=0)>0)
            varx = indcs1[0]
            indcs_list=list(group(varx))
            lenvar = len(indcs_list)
            j = 0
            while j < lenvar:
                var11 = indcs_list[j][0]
                var22 = indcs_list[j][1]
                var11 = var11 + xv1
                var22 = var22 + xv2
                morphed_row[:,var11:var22]=inv_notes_closed[red_lines[i]:red_lines[i+1],var11:var22]
                j = j+1
        arr_1 = np.ones((3,3))
        for i1 in range(0,3):
            if i1 == 0:
                notes3d_row[morphed_row>0,i1]=x
            else:
                notes3d_row[morphed_row>0,i1]=0
        notes[red_lines[i]:red_lines[i+1],:]=notes3d_row.copy()
        
        i = i+1
    return notes
```

### Key Note Detection

```python
def detectSharpsFlats(start,end,image,fimage,color):
    retImage = np.copy(fimage)
    staffRow = np.copy(image[start:end,:])
    currentFComponent = np.copy(fimage[start:end,:,:])
    colSumsComponent = np.sum(staffRow,axis = 0)
    miniSum = np.min(colSumsComponent)
    indices = np.where(colSumsComponent!=miniSum)[0]
    intervalsOfInterest = getConsecutiveIntervals(indices)[0:10]
    
    oSharps = 0
    oFlats = 0
    toign = 1
    lend = -1
    for intervalOfInterest in intervalsOfInterest:
        colStart = indices[intervalOfInterest[0]]
        colEnd = indices[intervalOfInterest[1]]
        if(colEnd-colStart)<=4:
            continue
        if toign>0:
            toign-=1
            continue
        if lend == -1:
            lend = colEnd
        else:
            if colStart-lend>=18:
                break
        
        workingArea = np.copy(staffRow[:,colStart:colEnd+1])
        fWorkingArea = np.copy(currentFComponent[:,colStart:colEnd+1,:])


        v_kernel = np.ones((25,1)).astype('uint8')
        erodedWorkingArea = cv2.erode(workingArea,v_kernel,iterations = 1)
        erodedWorkingArea = cv2.dilate(erodedWorkingArea,v_kernel,iterations = 1)

        workingAreaSum = np.sum(erodedWorkingArea,axis = 0)
        indics = np.where(workingAreaSum>0)[0]
        relevantIntervals = getConsecutiveIntervals(indics)
        
        numbSharps = 0
        numbFlats = 0
        if len(relevantIntervals) == 1:
            numbFlats = 1
            changes = 0
            for interval in relevantIntervals:
                a = indics[interval[0]]
                b = indics[interval[1]]
                a = (a+b)//2
                done = 0
                for i in range(erodedWorkingArea.shape[0]-1):
                    if erodedWorkingArea[i][a] != erodedWorkingArea[i+1][a]:
                        changes+=1
            if changes>2:
                numbFlats=0
        elif len(relevantIntervals) > 1:
            midFirst = 0
            olda = -1
            for interval in relevantIntervals:
                a = indics[interval[0]]
                b = indics[interval[1]]
                a = (a+b)//2
                done = 0
                for i in range(erodedWorkingArea.shape[0]):
                    if done == 0 and erodedWorkingArea[i][a] == 255:
                        done = i
                    elif done != 0 and erodedWorkingArea[i][a] == 0:
                        if midFirst == 0:
                            done = (done+i-1)//2
                        else:
                            done = i-1
                        break
                if midFirst == 0:
                    midFirst = done
                    olda = a
                else:
                    if (a-olda)>30 or (a-olda)<5:
                        break
                    if done<midFirst:
                        numbFlats = len(relevantIntervals)
                    else:
                        numbSharps = len(relevantIntervals)//2
                    break
        if (numbSharps+numbFlats)>0:
            for i in range(workingArea.shape[0]):
                for j in range(workingArea.shape[1]):
                    if(workingArea[i][j]>0):
                        for kl in range(3):
                            fWorkingArea[i][j][kl] = color[kl]
        else:
            break

        oSharps += numbSharps
        oFlats += numbFlats
#         else:
#             break
        currentFComponent[:,colStart:colEnd+1,:] = np.copy(fWorkingArea)
    retImage[start:end,:,:] = np.copy(currentFComponent)    
    return retImage.astype('uint8'),oSharps,oFlats
```

```python
numbSharpsRow = []
numbFlatsRow = []
for i in range(len(red_lines)-1):
    notes, numbSharps, numbFlats = detectSharpsFlats(red_lines[i],red_lines[i+1],inv_notes_closed,notes,[45,47,245])
    numbSharpsRow.append(numbSharps)
    numbFlatsRow.append(numbFlats)
print(notes.shape)
plt.figure(figsize=[10,10])
plt.title('Key Notes in Blue')
plt.imshow(notes)
plt.show()

print(numbSharpsRow)
print(numbFlatsRow)
```

### Dots and ties

```python
def getConnectedComponents(inv_notes_closed,i,j,label):
    inv_notes_closed[i][j] = label
    x= [i+1,i-1,i,i]
    y= [j,j,j-1,j+1]
    
    M = inv_notes_closed.shape[0]
    N = inv_notes_closed.shape[1]
    
    for k in range(4):
        if x[k]>=0 and x[k]<=M and y[k]>=0 and y[k]<N:
            if inv_notes_closed[x[k]][y[k]] == 1:
                getConnectedComponents(inv_notes_closed,x[k],y[k],label)

    return
```

```python
red2 = [0,75,37]
ties = [100,75,37]
workingAreaNotes = np.copy(inv_notes_closed)
workingAreaNotes[workingAreaNotes == 255] = 1
label = 2
mapperjMin = {}
mapperjMax = {}
mapperiMin = {}
mapperiMax = {}

for i in range(workingAreaNotes.shape[0]):
    for j in range(workingAreaNotes.shape[1]):
        if workingAreaNotes[i][j] == 1:
            getConnectedComponents(workingAreaNotes,i,j,label)
            mapperjMin[label] = j
            mapperjMax[label] = j
            mapperiMin[label] = i
            mapperiMax[label] = i
            label = label+1
        if workingAreaNotes[i][j]==0:
            continue
        if j<mapperjMin[workingAreaNotes[i][j]]:
            mapperjMin[workingAreaNotes[i][j]] = j
        if j>mapperjMax[workingAreaNotes[i][j]]:
            mapperjMax[workingAreaNotes[i][j]] = j
        if i<mapperiMin[workingAreaNotes[i][j]]:
            mapperiMin[workingAreaNotes[i][j]] = i
        if i>mapperiMax[workingAreaNotes[i][j]]:
            mapperiMax[workingAreaNotes[i][j]] = i

requiredLabels = []
requiredLabelsTies = []
ctr = 0
for x in mapperjMin:
    diffj = mapperjMax[x]-mapperjMin[x]
    diffi = mapperiMax[x]-mapperiMin[x]
    if diffj<15 and diffi<15 and diffj>=2 and diffi>=2:
        requiredLabels.append(x)
        ctr+=1
    elif diffj>=50 and diffj<=75 and diffi>=5 and diffi<15:
        requiredLabelsTies.append(x)

        
for label in requiredLabels:
    notes[workingAreaNotes==label] = np.copy(red2)
    
for label in requiredLabelsTies:
    notes[workingAreaNotes==label] = np.copy(ties)

plt.figure(figsize=[20,20])
plt.imshow(notes)
plt.show()
```

### Rest Detection

```python
miniDiffStave = np.min(np.diff(staves))

purple=[127,0,127]
z = np.copy(inv_notes_closed)

for i in range(len(red_lines)-1):
    invArea = inv_notes_closed[red_lines[i]:red_lines[i+1],:]
    notesLStaveRow = np.copy(notes[red_lines[i]:red_lines[i+1],:,:])
    
    colSumsComponent = np.sum(invArea,axis = 0)
    indices = np.where(colSumsComponent>0)[0]
    intervalsOfInterest = group(indices)
    
    notes_row = z[red_lines[i]:red_lines[i+1],:]
    for interval in intervalsOfInterest:
        colEnd = interval[1]
        colStart = interval[0]
        diff = colEnd-colStart
        checkArea = notes_row[:,colStart:colEnd+1]
        fWorkArea = np.copy(notesLStaveRow[:,colStart:colEnd+1,:])
        if len(np.unique(colSumsComponent[colStart:colEnd+1]))>2:
            continue
            
        rowSumsComponent = np.sum(checkArea,axis = 1)
        requiredRows = np.where(rowSumsComponent>0)[0]
        if len(requiredRows) == 0:
            continue
        nestedIntervals = group(requiredRows)
        
        tempoIntervals = []
        for ninterval in nestedIntervals:
            tempoIntervals.append(ninterval)
        nestedIntervals = tempoIntervals
        
        if(len(nestedIntervals)>1):
            continue
        ninterval = nestedIntervals[0]
        a = ninterval[0]
        b = ninterval[1]
        if (b-a)>=(miniDiffStave):
            continue
        if len(np.unique(rowSumsComponent[a:b+1]))>2:
            continue
        tempo = inv_notes_closed[red_lines[i]:red_lines[i+1],colStart:colEnd+1]
        for looper in range(3):
            fWorkArea[tempo>0,looper] = purple[looper]
        notesLStaveRow[:,colStart:colEnd+1,:] = np.copy(fWorkArea)
    
    notes[red_lines[i]:red_lines[i+1],:,:] = np.copy(notesLStaveRow)
plt.figure(figsize=[20,20])
plt.imshow(notes)
plt.show()
```

### Whole-Note and Half-Note Detection

```python
# circular structuring element used similar to quarter note detection 
images = []
titles = []
# royal_blue = [65,105,225]
# blue_violet	= [138,43,226]

royal_blue = [255,128,0]
blue_violet	= [204,153,255]

def whole_note(img):
    img_closed = cv2.morphologyEx(img, cv2.MORPH_CLOSE, circle_SE(7))
    img_closed_2 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, circle_SE(6))
    img = cv2.subtract(img_closed, img_closed_2)
    images.append(img)
    titles.append("whole note mid")
    img = cv2.morphologyEx(img, cv2.MORPH_OPEN, circle_SE(3))
    img = cv2.dilate(img, circle_SE(5))
    return img

def half_note(img):
    img_closed = cv2.morphologyEx(img, cv2.MORPH_CLOSE, circle_SE(5))
    img_closed = cv2.subtract(img_closed, img)
    # plt.imshow(img_closed, cmap='gray')
    images.append(img_closed)
    titles.append("half note mid")
    img_closed = cv2.morphologyEx(img_closed, cv2.MORPH_OPEN, circle_SE(3))
    img_closed = cv2.dilate(img_closed, circle_SE(5))
    return img_closed
```

```python
kernelMorph = circle_SE(int(0.3*(staves[1]-staves[0])))
for j in range(len(red_lines)-1):
    
    row = inv[red_lines[j]:red_lines[j+1],:]
    notes3d_row = notes[red_lines[j]:red_lines[j+1],:]
    
    sum = np.sum(row, axis=0)
    mode_val = stats.mode(sum[sum>0])
    mode_val = mode_val[0][0]
    indices = group(np.where(sum!=mode_val)[0])
    indices_list = list(indices)
    for i in range(5, len(indices_list)):
        if(j==0 and i<10):
            continue
        
        if(abs(indices_list[i][0]-indices_list[i][1])<10):
            continue;
            
        a, b = indices_list[i][0], indices_list[i][1]
        
        notes_row = inv_notes_closed[red_lines[j]:red_lines[j+1],a:b]
        notes3d_copy = notes3d_row[:,a:b]
        opened_row = cv2.morphologyEx(notes_row, cv2.MORPH_OPEN, kernelMorph)
        
        # images.append(opened_row)
        # titles.append("after opening")

        sum2 = np.sum(opened_row)
        if(sum2!=0):
            continue
        
        half_note_image = half_note(notes_row)
        images.append(half_note_image)
        titles.append("half note")
        whole_note_image = whole_note(notes_row)
        images.append(whole_note_image)
        titles.append("whole note")

        for k in range(3):
            notes3d_copy[half_note_image>0,k] = royal_blue[k]
            notes3d_copy[whole_note_image>0,k] = blue_violet[k]
        
        notes3d_row[:,a:b] = notes3d_copy.copy()
    
display_images(images, 3, 4, (20, 20), titles)
plt.figure(figsize=[10,10])
plt.imshow(notes)
plt.show()
```

```python
def Note(centroid_y, line_no):
    st_line = line_no*9
    stave_list=np.array(final_staves[st_line:st_line+9])
    a = np.diff(stave_list)
    a1=np.insert(a,0,[a[-1],a[-2]])
    a1=np.cumsum(a1)
    a1_f=stave_list[0]
    a2=np.append(a,[a[0],a[1]])
    a2=np.cumsum(a2)
    a2_f=a2
    a1_f-=np.flip(a1)
    a2_f+=stave_list[-1]
    final_stave_list=np.hstack((a1_f,stave_list,a2_f))
    line=np.argmin(np.array(abs(centroid_y-final_stave_list)))
    vals=['b','a','g','f','e','d','c','b','a','g','f','e','d','c','b','a','g','f','e','d','c','b','a','g','f','e','d','c','b',]
    str1 = ['f', 'c', 'g', 'd', 'a', 'e', 'b']
    str2 = ['b', 'e', 'a', 'd', 'g', 'c', 'f']
    value=vals[line]
    no_of_hashes, no_of_bs = numbSharpsRow[line_no], numbFlatsRow[line_no]
    
    if(no_of_hashes>0):
        for i in range(1, 8):
            if no_of_hashes>=i:
                if value==str1[i-1]:
                    value=str1[i-1]+'#'
    else:
        for i in range(1, 8):
            if no_of_bs>=i:
                if value==str2[i-1]:
                    value=str2[i-1]+'b'

    return value
```

```python
def Octave(centroid_x, centroid_y):
    red = [153,0,0]
    val=notes[centroid_x,centroid_y]
    color = [green, red, royal_blue, blue_violet, red2, purple, ties]
    oct_val = [4, 8, 2, 1, -1, 0, -10]
    
    for i in range(7):
        if(np.all(val==color[i])):
            return oct_val[i]

    return 100
```

```python
note_list=[]
octave_list=[]

for j in range(len(red_lines)-1):
    sum1=np.sum(inv[red_lines[j]:red_lines[j+1],:], axis=0)
    indcs=np.where(sum1!=stats.mode(sum1[sum1!=0])[0][0])
    indcs_list=group(indcs[0])
    indcs_list=list(indcs_list)
    notes_row=temp1[red_lines[j]:red_lines[j+1],:]
    
    for i in range(len(indcs_list)):
        a, b = indcs_list[i][0], indcs_list[i][1]
        if(a==b):
            continue
        nr = notes_row[:,a:b]
        if nr.shape[0] < 4 or nr.shape[1] < 4:
            continue
        (numLabels, labels, stats1, centroids) = cv2.connectedComponentsWithStats(nr, 4, cv2.CV_32S)
        sorted_indices=np.argsort(centroids[:,0])
        flag=1
        var1 = numLabels
        for k in range(var1):
            if k >= 0:
                var2 = max(stats1[:, cv2.CC_STAT_AREA])
            if(stats1[sorted_indices[k], cv2.CC_STAT_AREA]==var2):
                continue
            var3 = red_lines[j]
            x11,y11 = (sorted_indices[k],1)
            x22,y22 = (sorted_indices[k],0)
            var4 = centroids[x11,y11]
            var5 = var3 + var4
            var6 = centroids[x22,y22]
            x_mean=int(var5)
            y_mean=int(a+var6)
            note,octave=(Note(x_mean,j),Octave(x_mean, y_mean))
            lst_1 = [0,-100,-200]
            note_1 = ['r', 'r', 'r']
            nxt_1 = [4, 8, 16]

            if(octave==-1):
                octave_list[-1]=-octave_list[-1]

            if(octave==100 or octave==-1):
                continue

            for ii in range(3):
                if(octave==lst_1[ii]):
                    note=note_1[ii]
                    octave=nxt_1[ii]
            if octave_list:
                if octave_list[-1]==-10:
                    if(0>octave_list[-2]):
                        beat_1 = 4/(-octave_list[-2]) 
                        beat_1 += 4/(2*(-octave_list[-2]))
                    else:
                        beat_1=4/(octave_list[-2])
                    beat_2=4/octave
                    if((numLabels-1)!=k):
                        if(stats1[sorted_indices[k], cv2.CC_STAT_AREA]!=max(stats1[:, cv2.CC_STAT_AREA])):
                            x_mean, y_mean = int(red_lines[j]+centroids[sorted_indices[k],1]), int(a+centroids[sorted_indices[k],0])
                            if(Octave(x_mean, y_mean)==-1):
                                beat_2 = 4/octave 
                                beat_2 += 4/(2*octave)
                                flag=0
                    octave=4/(beat_1+beat_2)
                    for _ in range(2):
                        octave_list.pop()
                        note_list.pop()
            if(flag==0):
                flag=1
                continue
            if(100>stats1[sorted_indices[k], cv2.CC_STAT_AREA]):
                area_array=stats1[:, cv2.CC_STAT_AREA]
                k_val=np.argmax(area_array)
                lst_len = len(lst_1)
                area_array=np.delete(area_array,k_val)
                centroids=np.delete(centroids,k_val,0)
                nxt_len = len(nxt_1)
                temp = centroids[k,1]-centroids[:,1]
                dist=abs(temp)
                dist=dist[0!=dist]
                if(((staves[0]-staves[1])*1.5) < min(dist)):
                    continue
                
            note_list.append(note)
            octave_list.append(octave)
```
